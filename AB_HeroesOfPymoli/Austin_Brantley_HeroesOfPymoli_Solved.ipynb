## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
male_df = df.loc[df['Gender'] == 'Male',:]
fm_df = df.loc[df['Gender'] == 'Female',:]
o_df = df.loc[df['Gender'] == 'Other / Non-Disclosed',:]
```

```python
mc = len(np.unique(male_df['SN']))
fc = len(np.unique(fm_df['SN']))
oc = len(np.unique(o_df['SN']))
total_player_count = len(np.unique(df['SN']))
```

```python
count_dict = {'Total count': [mc, fc, oc]}
gendex=['Male','Female', 'Other / Non-Disclosed']
count_df = pd.DataFrame(data=count_dict, index = gendex)   
count_df['Percentage of Players'] = [mc/total_player_count*100, 
                                     fc/total_player_count*100, 
                                     oc/total_player_count*100]
count_df.index.name='Gender'

count_df['Percentage of Players'] = count_df['Percentage of Players'].map('{:,.2f}%'.format)
count_df.style.set_properties(**{'background-color': 'blue',
                           'color': 'yellow',
                            "text-align": "center"
                           })

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Retrieve the Item ID, Item Name, and Item Price columns
items = df[['Item ID', 'Item Name', 'Price']]
#Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
items_gb = items.groupby(['Item ID', 'Item Name'])
item_purch = items_gb['Price'].count()
item_revenue = items_gb['Price'].sum()
Price = item_revenue/item_purch

#Create a summary data frame to hold the results
item_df = pd.DataFrame({'Purchase Count':item_purch,
                        'Item Price':Price,
                        'Total Purchase Value':item_revenue
})


#Sort the purchase count column in descending order
sorted_df = item_df.sort_values(['Purchase Count'], ascending=False).head()

#Optional: give the displayed data cleaner formatting
style_sorted_df  = sorted_df.style.format({'Total Purchase Value':'${:,.2f}',
                                           'Item Price':'${:,.2f}'}).set_properties(**{'background-color': 'HoneyDew',
                                                                                                           'color': 'Indigo',
                                                                                                            "text-align": "center",
                                                                                                             })
#Display a preview of the summary data frame

style_sorted_df
```
